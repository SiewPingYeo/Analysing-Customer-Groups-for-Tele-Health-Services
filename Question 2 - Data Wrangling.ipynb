### 3. Question 2(a) % dropoff at each stage of the booking funnel by Month and by overall 

```python
# Create a function to calculate % change for each action stages 
def perc_change (df):
    pct= pd.DataFrame(df.groupby('action')['consult_id'].count())
    pct = pct.pct_change()
    pct.rename(columns = {'consult_id':'% dropout'}, inplace = True)
    return pct 
```

```python
# Calculate % dropout for October 
oct_pct = perc_change(oct)
oct_pct
```

```python
# Calculate % dropout for November 
nov_pct = perc_change(nov)
nov_pct
```

```python
# Calculate % dropout for December 
dec_pct = perc_change(dec)
dec_pct
```

```python
# Calculate % dropout for the year 
df_pct = perc_change(df)
df_pct
```

### 4. Question 2(b) Average time patients spent in each stage before successfully moving to the next stage

```python
# Create a function to calculate the average time spent for each stage 
def avg_time (df):
    df.sort_values(by = ['consult_id', 'action'], inplace = True)
    df['time_in_status'] = df.groupby('consult_id')['timestamp'].diff()
    final = pd.DataFrame(df.groupby('action')['time_in_status'].mean())
    return final 
```

```python
# Calculate average time spent for Oct 
oct_avgtime = avg_time(oct)
```

```python
# Calculate average time spent for Nov
nov_avgtime = avg_time(nov)
```

```python
# Calculate average time spent for Dec
dec_avgtime = avg_time(dec)
```

```python
# Calculate average time spent for overall 
df_avgtime = avg_time(df)
```

```python
# Convert timestamp to text 
dataframes = [oct_avgtime, nov_avgtime, dec_avgtime, df_avgtime]

for i in dataframes:
    i['time_in_status'] = i['time_in_status'].astype(str)
```
